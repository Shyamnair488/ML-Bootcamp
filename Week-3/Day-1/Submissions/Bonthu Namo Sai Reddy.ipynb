```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/ML-Bootcamp/Week-3/Day-1/Dataset/SMSSpamCollection.txt", sep="\t", names = ['label','text'])
df
```

```python
df.shape
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer
nltk.download('stopwords')
nltk.download('punkt')
```

```python
sn = SnowballStemmer("english")
stop = set(stopwords.words("english"))
```

```python
def clean_text(sms):
  sms = sms.lower()
  sms = re.sub("[^a-z]",' ',sms)
  sms = nltk.word_tokenize(sms)
  sms = [t for t in sms if len(t)>1]
  sms = [sn.stem(word) for word in sms if word not in stop]
  sms = ' '.join(sms)
  return sms
```

```python
clean_text("Get this stuff for playing free ? just Rs.0")
```

```python
df['clean_text'] = df['text'].apply(clean_text)
```

```python
df.head()
```

```python
from wordcloud import WordCloud
```

```python
hamdata = df[df['label'] == 'ham']
hamdata = hamdata['clean_text']
```

```python
def wordCloud(data):
  words = ' '.join(data)
  wc = WordCloud(background_color='white')
  wc = wc.generate(words)

  plt.figure(figsize = (10,8))
  plt.imshow(wc)
  plt.axis("off")
  plt.show()
```

```python
wordCloud(hamdata)
```

```python
spamdata = df[df['label'] == 'spam']
spamdata = spamdata['clean_text']
```

```python
wordCloud(spamdata)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features = 5000)
```

```python
X = cv.fit_transform(df['clean_text']).toarray()
```

```python
X.shape
```

```python
y = pd.get_dummies(df['label'])
y = y['spam'].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_test,y_pred
```

```python
pd.DataFrame({'Actual' : y_test, 'predicted': y_pred}).head(10)
```

```python
print("Accuracy on Training Data :")
print(model.score(X_train,y_train))
```

```python
print("Acuuracy on Testing Data:")
print(model.score(X_test,y_test))
```
