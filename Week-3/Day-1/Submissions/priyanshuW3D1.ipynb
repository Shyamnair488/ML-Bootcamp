```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/ML-Bootcamp/Week-3/Day-1/Dataset/SMSSpamCollection.txt", sep="\t", names = ['label','text'])
data
```

# Data Cleaning

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer
nltk.download('stopwords')
nltk.download('punkt')
```

```python
sn = SnowballStemmer("english")
stop = set(stopwords.words("english"))
```

```python
def clean_text(sms):
    sms = sms.lower()
    sms = re.sub("[^a-z]", ' ', sms)
    sms = nltk.word_tokenize(sms)
    sms = [t for t in sms if len(t)>1]
    sms = [sn.stem(word) for word in sms if word not in stop]
    sms = ' '.join(sms)
    return sms
```

```python
clean_text("This is a Sample Test to CHECK and remove ? @ #* $ 123 i playing")
```

```python
data['clean_text'] = data['text'].apply(clean_text)
```

```python
data.head()
```

# WordCloud

```python
from wordcloud import WordCloud
```

```python
ham_data = data[data['label'] == 'ham']
ham_data = ham_data['clean_text']
```

```python
def wordcloud(data):
    words = ' '.join(data)
    wc = WordCloud(background_color='white')
    wc = wc.generate(words)

    plt.figure(figsize = (10, 8))
    plt.imshow(wc)
    plt.axis("off")
    plt.show()
```

# visuals for ham data

```python
wordcloud(ham_data)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features= 5000)
cv
```

```python
x = cv.fit_transform(data['clean_text']).toarray()
x
```

```python
x.shape
```

```python
y = pd.get_dummies(data['label'])
y
```

```python
y = y['spam'].values
y
```

# Model building

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.33, random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
```

```python
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
Y_test, Y_pred
```

```python
pd.DataFrame({'Actual' : Y_test, 'predicted': Y_pred}).head(10)
```

```python
print("Accuracy on Training Data: ")
print(model.score(X_train, Y_train))
```

```python
print("Accuracy on Training Data: ")
print(model.score(X_test, Y_test))
```
